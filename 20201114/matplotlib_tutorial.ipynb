```python
import matplotlib.pyplot as plt
import numpy as np
#print(matplotlib.__version__)
```

```python
x_list = [0,1,2,3,4,5,6]
y_list = [0,5,2,1,3,5,6]
x = np.array(x_list)
y = np.array(y_list)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'o')
plt.show()
```

```python
plt.plot(x,y,'x')
plt.show()
```

```python
plt.plot(x,y,marker="x")
plt.show()
```

```python
plt.plot(x,y,'o:')
plt.show()
```

```python
plt.plot(x,y,'o--')
plt.show()
```

```python
plt.plot(x,y,'o-.g')
plt.show()
```

```python
plt.plot(x,y,'o-b')
plt.show()
```

```python
x2_list = [0,1,2,3,4,5,6]
y2_list = [5,6,7,3,3,1,2]
x2 = np.array(x2_list)
y2 = np.array(y2_list)
```

```python
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

```python
# plot : 1
plt.subplot(1,2,1)
plt.plot(x,y)

# plot : 2
plt.subplot(1,2,2)
plt.plot(x2,y2)

plt.show()
```

```python
# plot : 1
plt.subplot(2,1,1)
plt.plot(x,y)

# plot : 2
plt.subplot(2,1,2)
plt.plot(x2,y2)

plt.show()
```

```python
# plot : 1
#  (,,)
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("fig.01")

# plot : 2
plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.title("fig.02")

# plot : 3
plt.subplot(2,2,3)
plt.plot(x,y)
plt.title("fig.03")

# plot : 4
plt.subplot(2,2,4)
plt.plot(x2,y2)
plt.title("fig.04")

plt.suptitle('sample figure')
plt.show()
```

```python
t = np.arange(0.,5.,0.2)
plt.plot(t,t,'r--')
plt.plot(t,t**2,'bs')
plt.plot(t,t**3,'g^')
plt.show()
```
