```python
fig, axs = plt.subplots(1,2,figsize=(12,6))
explode = (0,0.1)
labels = ['dead','alive']
df['survived'].value_counts().plot.pie(explode=explode,labels=labels,autopct='%1.1f%%',shadow=True,startangle=90,ax=axs[0])
sns.countplot(df['survived'],ax=axs[1])

plt.show()
```
