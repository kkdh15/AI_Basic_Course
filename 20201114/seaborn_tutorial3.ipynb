```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')
print(type(titanic))
```

```python
titanic.head()
```

```python
grid = sns.FacetGrid(data=titanic, col='who',row='survived')
```

```python
grid.map(plt.hist,'age')
plt.show()
```

```python
titanic_pair = titanic[['age','pclass','fare']]
```

```python
sns.pairplot(titanic_pair,size=10)
plt.show()
```
