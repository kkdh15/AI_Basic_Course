```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.set_style('darkgrid')
figure = plt.figure(figsize=(15,5))

fig_1 = figure.add_subplot(1,2,1)
fig_2 = figure.add_subplot(1,2,2)
sns.regplot(data = titanic, x= 'age',y= 'fare',ax = fig_1)
sns.regplot(data = titanic, x= 'age',y= 'fare',ax = fig_2,fit_reg=False)
plt.show()
```

```python
sns.regplot(data = titanic, x= 'age',y= 'fare',ax = fig_1)
```

```python
sns.lmplot(x='age',y='fare',data=titanic)
```

```python
# dsiplot : 값의 분포를 확인
print(titanic['fare'].head())
sns.displot(titanic['fare'])
plt.show()
```

```python
sns.displot(titanic['fare'],kind ='kde')
plt.show()
```

```python
sns.kdeplot(titanic['fare'])
plt.show()
```

```python
print(titanic.head())
sns.lmplot(x='age',y='fare',hue='sex',data = titanic)
plt.show()
```

```python
sns.lmplot(x='age',y='fare',col='sex',hue='sex',data=titanic,size=12,palette='Set1')
plt.show()
```

```python
sns.pairplot(titanic,size = 4,vars = ['age','fare'])
plt.show()
```

```python
sns.set_theme(style='ticks')

sns.pairplot(titanic,size=4,vars = ['age','fare'],hue='sex')
plt.show()
```
