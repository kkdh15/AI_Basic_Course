```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')
print(type(titanic))
```

```python
sns.set_style('darkgrid')
titanic.head()
```

```python
table = titanic.pivot_table(index=['sex'],columns=['class'],aggfunc='size')
table.head()
```

```python
cmap = sns.diverging_palette(230,20,as_cmap=True)
sns.set_theme(style='white')
sns.heatmap(table,
            annot=True,
            fmt='d',
            cbar=True,
            lineWidths=1,
            linecolor='white',
            cmap=cmap)
plt.show()
```

```python
#jointplot
sns.jointplot(data=titanic,x='age',y='fare',height=10)
plt.show()
sns.jointplot(data=titanic,x='age',y='fare',height=10,kind='kde')
plt.show()
sns.jointplot(data=titanic,x='age',y='fare',hue='sex',height=10,kind='reg')
plt.show()
```
